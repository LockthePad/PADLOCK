```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# 저장할 폴더 설정
save_dir = "C:/Users/SSAFY/Desktop/padlock/board_img"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Selenium WebDriver 설정
driver = webdriver.Chrome()
url = "https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query=%ED%95%99%EA%B5%90++%EC%B9%A0%ED%8C%90"
driver.get(url)

# 페이지 로드 대기
time.sleep(2)

last_height = driver.execute_script("return document.body.scrollHeight")
img_urls = set() 

while True:
    images = driver.find_elements(By.TAG_NAME, 'img')
    
    for img in images:
        img_url = img.get_attribute('src') or img.get_attribute('data-src')
        if img_url and img_url.startswith('http'):
            img_urls.add(img_url)
    
    # 스크롤 내리기
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

for i, img_url in enumerate(img_urls):
    try:
        img_data = requests.get(img_url).content
        with open(os.path.join(save_dir, f'image_{i}.jpg'), 'wb') as f:
            f.write(img_data)
        print(f"Image {i} saved.")
    except Exception as e:
        print(f"Failed to save image {i}: {e}")

driver.quit()
```
